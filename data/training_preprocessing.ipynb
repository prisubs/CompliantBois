```python
import pandas as pd
import numpy as np
```

```python
file_names_original = ["AAPL.csv", "AMZN.csv", "BABA.csv", "CRM.csv", "FB.csv", "GOOGL.csv", 
                       "MSFT.csv", "NFLX.csv", "TSLA.csv", "WMT.csv"]

def read_clean(filename):
    DESIRED = ["Date", "Open", "Close", "High", "Low", "Volume", "Average", "Ticker"]
    df = pd.read_csv(filename)
    df = df[DESIRED]
    return df
```

```python
for file in file_names_original:
    df = pd.read_csv("original/" + file)
    df["Average"] = (df["High"] + df["Low"]) / 2
    df["Ticker"] = file[0:-4]
    df.to_csv(file)
```

```python
for file in file_names_original:
    display(read_clean(file).tail())
```
